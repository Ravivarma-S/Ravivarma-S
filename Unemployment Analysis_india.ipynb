```python
#removes the row where all the values are null
dfval = dfval.dropna(how='all')
dfval
```

Estimated Amount of Employment in India at each state

```python
plt.figure(figsize=(10,12))
plt.title("Estimated Employement rate")
sns.histplot(x = ' Estimated Employed' , hue = 'Region' , data = dfval)
plt.show()
```

Estimated Amount of unemployment in India at each state

```python
plt.figure(figsize=(10,12))
plt.title("Estimated unemployement rate")
sns.histplot(x = ' Estimated Unemployment Rate (%)' , hue = 'Region' , data = dfval)
plt.show()
```

Now we will create a dashboard for seeing the unemployment rate in different regions of india

```python
import plotly.express as px
unemployement = dfval[['Region',' Estimated Unemployment Rate (%)','Area']]
figure = px.sunburst(dfval,path=['Region','Area'],values = ' Estimated Unemployment Rate (%)',width =1000,height=1000,title = "Unemployement in India")
figure.show()
```
